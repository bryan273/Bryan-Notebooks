## Recursion

```
public static int factorialRec(int n) {
  if (n == 0)
    return 1;
  else
    return n * factorialRec(n-1);
}

factorialRec(5);
// it will throw java.lang.StackOverflowError kalo gada base case
```

```
public static int factorialRec(int n) {
  return n == 0 ? 1 : n * factorialRec(n-1);
  // if-else condition ? trueCase : falseCase
}

factorialRec(5);
```

## Class code concept

```
class Circle {
  // Data field
  double radius = 1.0;  

  // Constructor with no args -> default constructor
  // Must defined if we want to add constructor with args
  // If dont , doesnt need to declare, already automatically
  Circle(){
  }

  // Constructor with args
  Circle(double newRadius){
    radius = newRadius;
  }

  // Method
  double getArea(){
    return radius*radius* 3.14;
  }
}

Circle circle1 = new Circle();
Circle circle2 = new Circle(25.0);

System.out.println(circle1.radius);
System.out.println(circle2.radius);
System.out.println(circle1.getArea());
System.out.println(circle2.getArea());
```

```
public class Box{
  String name; // if we dont define value , default null
  double a; // default 0
  boolean b; // default false
  char c; // default '\u0000'
  Scanner d; // default value null , null no special value , NullPointerException
}

Box box1 = new Box();
box1.a;
```

```
class Circle {
  double radius;

  // default value for this
  Circle(){
    this(5.0);
  }

  // this keyword for self point object
  Circle(double radius){
    this.radius = radius;
  }

  double getArea(){
    return this.radius * this.radius* 3.14;
  }
}

Circle circle1 = new Circle(10.0);
Circle circle2 = new Circle();
System.out.println(circle1.getArea());
System.out.println(circle2.getArea());
```

## toString & equals

```
public class Cube{
  String color;
  double length;

  public Cube() {
    this.color = "White";
    this.length = 1.0;
  }

  public Cube(String color, double length) {
    this.color = "White";
    this.length = 1.0;
  }  

  public String toString() {
    return String.format("Cube with length = %.2f and color = %s", this.length, this.color);
  }

  public boolean equals(Cube otherCube) {
    return this.color == otherCube.color && this.length == otherCube.length;
  }
}

Cube cube1 = new Cube();
Cube cube2 = new Cube("White", 1.0);
System.out.println(cube2);
System.out.println(cube1.equals(cube2));
```

## Immutable Class

If the contents of an object cannot be changed once the object is
created, the object is called an immutable object and its class is called
an immutable class. 

For a class to be immutable:
* All data fields must be private.
* There canât be any mutator methods for data fields.
* No accessor methods can return a reference to a data field that is mutable.


```
public class Student {
  private int id;
  private BirthDate birthDate;
  
  public Student(int ssn,int year, int month, int day) {
    id = ssn;
    birthDate = new BirthDate(year, month, day);
  }

  public int getId() {
    return id;
  }

  public BirthDate getBirthDate() {
    return birthDate;
  }
}
```

Javaâs layout managers provide a level of abstraction to automatically map your

* Grid Layout -> creates specified number of rows and columns.
* Border Layout -> Creates a Border manager divides the container into five areas: East, South,

```
import java.awt.FlowLayout;

public class ShowFlowLayout extends JFrame {
  public ShowFlowLayout() {
    // aligned left with horizontal gap 10
    // and vertical gap 20 between components
    setLayout(new FlowLayout(FlowLayout.LEFT, 10, 20));

    // 3 rows, 2 columns, and gaps 5 between
    // components horizontally and vertically
    setLayout(new GridLayout(3, 2, 5, 5));

    // Add labels and text fields to the frame
    add(new JLabel("First Name"));
    add(new JTextField(8)); // Size of field
    add(new JLabel("MI"));
    add(new JTextField(1));
    add(new JLabel("Last Name"));
    add(new JTextField(8));

    // horizontal gap 5 and vertical gap 10
    setLayout(new BorderLayout(5, 10));
    // Add buttons to the frame
    add(new JButton("East"), BorderLayout.EAST);
    add(new JButton("South"), BorderLayout.SOUTH);
    add(new JButton("West"), BorderLayout.WEST);
    add(new JButton("North"), BorderLayout.NORTH);
    add(new JButton("Center"), BorderLayout.CENTER);
  }

  public static void main(String[] args) {
    ShowFlowLayout frame = new ShowFlowLayout();
    frame.setTitle("ShowFlowLayout");
    frame.setSize(200, 200);
    frame.setLocationRelativeTo(null); // Center the frame
    frame.setDefaultCloseOperation(JFrame.EXIT_ON_CLOSE);
    frame.setVisible(true);
  }
}
```

## ArrayList

* ArrayList is a dynamic-size array
* Can adjust the capacity as needed
* Provides methods for managing the array, i.e,. adding an element, removing an element

```
import java.util.ArrayList;

ArrayList<String> names = new ArrayList<String>(); // initialize array list

names.add("priyanka"); // add object
names.add("deyuna");
names.add(1, "bryan"); // add object at index

String name = names.get(1); // get object
System.out.println(name);

names.set(1, "Carolyn"); // set object
System.out.println(names);

names.remove(1); // remove index
System.out.println(names);

names.remove("deyuna"); // remove object
System.out.println(names);

names.clear(); // clear object
System.out.println(names);
```

* ArrayList stores a collection of "objects" contiguously
* Stores Objects â uses wrapper classes for primitive values

```
ArrayList<Integer> arrayList = new ArrayList<Integer>();
arrayList.add(25); // Autoboxing because ArrayList stores only objects
int num = arrayList.get(0); // unboxing since get returns an Integer object
```

## Generics

* Gives us the capability to parameterize freely the types.
* You can specify allowable types of objects that the class or method may work with.
* If you attempt to use the class or method with an incompatible object, a compile error occurs.
* You can also define a class or a method with generic types.


```
public static <T> void print(ArrayList<T> list) {
  for(int i = 0; i < list.size(); i++) {
    System.out.println(list.get(i));
  }
}

// you can specify T with any Object , rather than 
// creating 3 different functions with different types of parameter objects
```

```
public static <T> void hello(T obj) {
  System.out.println("Hello, " + obj);
}

hello("Hello");
hello(new Integer(5));
String[] strArr = {"1", "2", "3"};
hello(new ArrayList<String>(Arrays.asList(strArr)));
```

```
// without generics

ArrayList ageList = new ArrayList(); // called raw type
ageList.add(new Integer(46));
ageList.add("50");
Integer sum = new Integer(0);

for (Object age : ageList) {
  sum = sum + ((Integer) age);
}
```

```
// Max1.java: Find a maximum object
public class Max1 {
  /** Return the maximum between two objects */
  public static <E extends Comparable<E>> E max(E o1, E o2) {
  if (o1.compareTo(o2) > 0)
    return o1;
  else
    return o2;
  }
}

// Max1.max(1,"abcd") compile error
Max1.max(1,2)
```

* Java does not make a distinction between interfaces and classes when defining bounds (inside < >) for generic type parameters
* Extends is used for both interface and class
types which makes expressing the bound direction more streamlined.


```
// We ensure the generic type to be bounded to subclasses of GeometricObject
public static <E extends GeometricObject> boolean equal(E object1, E object2) {
  return object1.getArea() == object2.getArea();
}
```

* It is important to note that a generic class is shared by all its instances regardless
of its actual generic type.
* Although ArrayList\<String> and ArrayList\<Integer> are two types,
but there is only one class ArrayList.

* Set and List are subinterfaces of Collection

* A list stores elements in a sequential order and allows the user to specify where
the element is stored. The user can access the elements by index.
* ArrayList and LinkedList class are concrete implementations of the
List interface.  
* If you need to support random access through an index without inserting or
removing elements from any place other than the end, ArrayList offers the
most efficient collection.
* If your application requires the insertion or deletion of elements for first and last place, you should choose LinkedList.

* A stack is a collection of last-in-first-out (LIFO) stack of objects. The elements are
accessed only from the top of the stack. You can retrieve, insert, or remove an
element only from the top of the stack.

```
Stack<Integer> stack = new Stack<Integer>();
stack.push(10);
stack.push(20);
stack.push(30);
stack.push(40);

System.out.println(stack.peek());
while(!stack.isEmpty()) {
  System.out.println(stack.pop());
}
```
