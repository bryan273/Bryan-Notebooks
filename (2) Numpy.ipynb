## Numpy reshape

```python
print('Array:\n',x,'\n')
print('Reshape 3x2:\n',x.reshape(3,2),'\n')

# 3 fungsi ini untuk flatten data (1D)
print('Flatten:',x.reshape(6,),'')
print('Flatten:',x.reshape(-1,),'') # -1 -> max representation / reshape(6,)
print('Flatten:',x.flatten(),'')
print('Flatten:',x.ravel(),'\n')

print('Reshape 6x1:\n',x.reshape(-1,1),'\n') # reshape(6,1) , max representation -1 -> 6 (mempertahankan size = 6 (6x1) )
print('Reshape 1x6:\n',x.reshape(1,-1),'\n') # reshape(1,6) , max representation -1 -> 6 (mempertahankan size = 6 (1x6) )
print('Reshape 3x2:\n',x.reshape(-1,2),'\n') # reshape(3,2) , max representation -1 -> 3 (mempertahankan size = 6 (3x2) )

# Contoh , kita mempunyai data yang kita tidak tahu sizenya berapa
# Tetapi kita ingin me-reshape supaya mempunyai 2 kolom , maka kita bisa melakukan (-1,2)
# -1 akan menyesuaikan dengan angka yang tersedia untuk mempertahankan size 
```

## Masking

```python
x = np.array([[1,2,3],[4,5,6]])
print('Array:\n',x,'\n')
print('Boolean value more than 2:\n',x > 2,'\n') # Ini masking namanya
print('Boolean for odd value:\n',x%2==1,'\n')
```

```python
# Masking bisa di index
# Kalo True , data di posisi itu diambil
print('Array:\n',x,'\n')
print('Value more than 2:\n',x[x>2],'\n')

mask = np.array([[False, False,  False] , [False,  False,  True]])
print('Value in the True index:',x[mask],'\n')

# True False beda dengan indexing 0 dan 1
# Kalo ini dia bakal ambil index [[0]] atau [[1]] , bukan nilai yang di posisi 1
idx = np.array([[0, 0,  0] , [0,  0,  1]])
print('Value in the idx index:\n',x[idx])
```

```python
print('Value more than 2 and even:\n',x[(x>2)&(x%2==0)],'\n')
print('Value more than 2 or even:\n',x[(x>2)|(x%2==0)],'\n')
```

## List vs Numpy array

```python
import time
 
now = time.time() # Get time now

size = 100000
list1 = range(size)
list2 = range(size)
 
result = [(a * b) for a, b in zip(list1, list2)] # Multiplication for list in each index
 
later = time.time() # Get time after operation

print("List :",
      round((later - now)*1e3),
      "mili-seconds")
#1e3 = 10^3
```

```python
now = time.time()

size = 100000
arr1 = np.arange(size) 
arr2 = np.arange(size)

result = arr1 * arr2 # Multiplication for numpy in each index
 
later = time.time()

print("Numpy Array :",
      round((later - now)*1e3),
      "mili-seconds")
```

## Numpy argmax

```python
# np.argmax return index dengan value terbesar

x = np.array([[1,3,5],[6,4,5],[1,2,5]])

print('data :\n',x,'\n')
print('maximum value is in index :',np.argmax(x)) # value terbesar dengan index yang di flatten
print('maximum value in each column is in index :',np.argmax(x, axis=0)) # value terbesar dari index per kolom, yg di return index list dimana nilai terbesar itu ada
print('maximum value in each row is in index :',np.argmax(x, axis=1)) # value terbesar dari index per baris, yg di return index object dalam 1 list
```

**Soal 1**

Misalkan anda memiliki data [5,6,8,2,3,9,6,11,20,1,2,8,2,1,19]. Sebagai seorang statistika , anda diminta untuk mencari insight dari data tersebut.

Pertanyaan:
1.   Carilah nilai minimum , maksimum , rata rata , median , std , varian dari data tersebut dengan pembulatan 2 desimal. 
2.   Carilah apakah data tersebut memiliki outlier ? Jika ada sebutkan nilai mana yang merupakan outlier.

Note: 
- Pakailah rumus outlier untuk menentukan data mana yang merupakan outlier. 
- Gunakan np.quantile(data, 0.25)) untuk menghitung Q1 dan np.quantile(data, 0.75)) untuk Q3

```python
# Tulis jawaban anda disini
data = [5,6,8,2,3,9,6,11,20,1,2,8,2,1,19]
```

**Soal 2**

Misalkan anda memiliki suatu data baru yang belum anda ketahui detailnya. Tugas anda adalah mencari info dari data tersebut.

Pertanyaan:

1. Carilah jumlah dimensi , shape , size data tersebut
2. Lakukanlah scaling terhadap data tersebut dengan min max scaler
3. Carilah persentase data yang tersebar di sekitar mean




# Solusi

```python
import numpy as np

data = [5,6,8,2,3,9,6,11,20,1,2,8,2,1,19]
data = np.array(data)

print('max:', data.max().round(2))
print('min:', data.min().round(2))
print('mean:', np.mean(data).round(2))
print('med:', np.median(data).round(2))
print('std:', np.std(data).round(2))
print('var:', np.var(data).round(2))

Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)

IQR = Q3-Q1
lower_outlier = Q1-1.5*IQR
upper_outlier = Q3+1.5*IQR

print('Q1:', Q1)
print('Q3:', Q3)
print('lower outlier:', lower_outlier)
print('upper outlier:', upper_outlier)
print('outlier:', any((data<lower_outlier) | (data>upper_outlier)))
print('outlier value:', data[(data<lower_outlier) | (data>upper_outlier)])
```
